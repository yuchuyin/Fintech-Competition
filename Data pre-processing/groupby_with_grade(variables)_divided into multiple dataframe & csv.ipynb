```python
for i in range(0,len(grade_list)):
    
    locals()['ALL_'+str(grade_list[i])]=group.get_group(grade_list[i]) #
    locals()['X_num'+str(grade_list[i])]=locals()['ALL_'+str(grade_list[i])][X_num_use_var.columns]
    locals()['X_char'+str(grade_list[i])]=locals()['ALL_'+str(grade_list[i])][X_char_use_var.columns]
    locals()['y_all'+str(grade_list[i])]=locals()['ALL_'+str(grade_list[i])][['Unnamed: 0','loan_status']]
    
    
    # data中沒有 missing value 的部分(觀測值)
    filter=locals()['X_num'+str(grade_list[i])].isnull().sum(axis=1)==0
    locals()['X_num'+str(grade_list[i])+'_rowwithoutna']=locals()['X_num'+str(grade_list[i])][filter] # 含有 NA的 觀測值分
    print(locals()['X_num'+str(grade_list[i])+'_rowwithoutna'].shape)

    # data中有 missing value 的部分(觀測值)
    filter=locals()['X_num'+str(grade_list[i])].isnull().sum(axis=1)!=0
    locals()['X_num'+str(grade_list[i])+'_rowwithna']=locals()['X_num'+str(grade_list[i])][filter] # 含有 NA的 觀測值分
    print(locals()['X_num'+str(grade_list[i])+'_rowwithna'].shape)
    
   
    # 補植-num
    imp = IterativeImputer(max_iter=10, random_state=0)
    imp.fit(locals()['X_num'+str(grade_list[i])]) # 整筆 _X_train_num
                                        
                                                   # 要補值的部分
    X_num_rowwithna_imp=pd.DataFrame(imp.transform(locals()['X_num'+str(grade_list[i])+'_rowwithna']),index=locals()['X_num'+str(grade_list[i])+'_rowwithna'].index,columns=locals()['X_num'+str(grade_list[i])+'_rowwithna'].columns)
    print(X_num_rowwithna_imp.shape)


    # 上下合併  補完後的部分 和 沒有 missing value(觀測直) 的部分合併
    locals()['X_num_imp'+str(grade_list[i])]=pd.concat([locals()['X_num'+str(grade_list[i])+'_rowwithoutna'],X_num_rowwithna_imp],axis=0)
    
    # 確保每個觀測直的位子和原本一樣，沒有打亂。前半是train 
    locals()['X_num_imp'+str(grade_list[i])]=locals()['X_num_imp'+str(grade_list[i])].sort_index()  
    print(locals()['X_num_imp'+str(grade_list[i])].shape)
    
    
    
    
    # train test 一起做 pca
    X_num_use_var_id = locals()['X_num_imp'+str(grade_list[i])]['Unnamed: 0']

    # 拿掉 id
    locals()['X_num_imp'+str(grade_list[i])]= locals()['X_num_imp'+str(grade_list[i])].drop(columns=['Unnamed: 0'])

    # 标准化数据
    ss = StandardScaler()
    locals()['X_num_imp_std'+str(grade_list[i])]= ss.fit_transform(locals()['X_num_imp'+str(grade_list[i])])
    
    ###  PCA
    pca=PCA(n_components=0.9,copy=True, whiten=False)
    locals()['X_num_imp_pca'+str(grade_list[i])]=pca.fit_transform(locals()['X_num_imp_std'+str(grade_list[i])])

    #print(pca.explained_variance_ratio_) 
    #print(pca.explained_variance_)
    
    print(pca.n_components_)
    
    # 查看降维后的数据
    locals()['X_num_imp_pca'+str(grade_list[i])] = pd.DataFrame(data= locals()['X_num_imp_pca'+str(grade_list[i])])
 
    # 加回 'Unnamed: 0'
    locals()['X_num_imp_pca'+str(grade_list[i])]['Unnamed: 0']= X_num_use_var_id.reset_index()['Unnamed: 0']

    
    
    
    
    # CHAR 
    # 把 na 當成一類
    
    var= locals()['X_char'+str(grade_list[i])].columns.tolist()
    for j in range(0,len(var)): 
         locals()['X_char'+str(grade_list[i])][str(var[j])]=locals()['X_char'+str(grade_list[i])][str(var[j])].fillna('unknown')
    
    
    #轉dummy
    
    X_id = locals()['X_char'+str(grade_list[i])]['Unnamed: 0']
    
    # 拿掉 'Unnamed: 0.1'
    locals()['X_char'+str(grade_list[i])]=locals()['X_char'+str(grade_list[i])].drop(columns=['Unnamed: 0'])
    locals()['X_char_dummy'+str(grade_list[i])]=pd.get_dummies(locals()['X_char'+str(grade_list[i])],drop_first=True)
    
    locals()['X_char_dummy'+str(grade_list[i])]['Unnamed: 0']=X_id
    
    
    os.makedirs(path+'/grade'+'/'+str(grade_list[i])) #建一個資料夾
    
    locals()['ALL_'+str(grade_list[i])].to_csv(path+'/grade'+'/'+str(grade_list[i])+'/'+ 'ALL_'+str(grade_list[i])+'.csv')
    locals()['X_num'+str(grade_list[i])].to_csv(path+'/grade'+'/'+str(grade_list[i])+'/'+ 'X_num'+str(grade_list[i])+'.csv')
    locals()['X_char'+str(grade_list[i])].to_csv(path+'/grade'+'/'+str(grade_list[i])+'/'+ 'X_char'+str(grade_list[i])+'.csv')
    locals()['y_all'+str(grade_list[i])].to_csv(path+'/grade'+'/'+str(grade_list[i])+'/'+ 'y_all'+str(grade_list[i])+'.csv')
    locals()['X_num_imp'+str(grade_list[i])].to_csv(path+'/grade'+'/'+str(grade_list[i])+'/'+ 'X_num_imp'+str(grade_list[i])+'.csv')
    locals()['X_char_dummy'+str(grade_list[i])].to_csv(path+'/grade'+'/'+str(grade_list[i])+'/'+ 'X_char_dummy'+str(grade_list[i])+'.csv')
```

```python
# A


y_trainA=pd.read_csv(path+'/grade/A/y_trainA.csv')
y_testA=pd.read_csv(path+'/grade/A/y_testA.csv')

X_testA=pd.read_csv(path+'/grade/A/X_testA.csv',index_col=0)
X_trainA=pd.read_csv(path+'/grade/A/X_trainA.csv',index_col=0)



# 處理 y
le = LabelEncoder()
le.fit(y_trainA['loan_status'].unique())  
y_trainA['ans']=le.transform(y_trainA['loan_status'])

#drop id

X_trainA.drop(columns=['Unnamed: 0.1'])
X_testA.drop(columns=['Unnamed: 0.1'])

##

gsc = GridSearchCV(estimator=RandomForestClassifier(),param_grid={
            'max_features':['sqrt','auto','log2',0.5,0.7],
            'max_depth': range(1,6),
            'n_estimators': range(1,101,10)},
             cv=10, scoring='f1',verbose=0,n_jobs=-1,)

grid_result = gsc.fit(X_trainA,y_trainA.ans)
best_params = grid_result.best_params_
print(best_params)

rfr = RandomForestClassifier(
    max_depth=best_params["max_depth"], 
    n_estimators=best_params["n_estimators"],
    max_features=best_params['max_features'],
    random_state=False, verbose=False)

scores = cross_val_score(rfr, X_trainA,y_trainA.ans, cv=10, scoring='f1')
print(scores)

rfr.fit(X_trainA,y_trainA.ans)


# save model
#儲存Model(注:save資料夾要預先建立，否則會報錯)

with open( path+'\grade\A\RF_A.pickle', 'wb') as f:
    pickle.dump(rfr, f)

#讀取Model
#with open('save/clf.pickle', 'rb') as f:
#    clf2 = pickle.load(f)
    #測試讀取後的Model
#   print(clf2.predict(X[0:1]))

## predict_y

rf_1_pred=rfr.predict(X_testA)

y_ans = pd.DataFrame(rf_1_pred,columns=['predict'],index=y_testA.index)
y_output=pd.merge(y_testA,y_ans,left_index=True,right_index=True)
y_output.to_csv(path+'\grade\A\y_output_rf_A.csv')
```
