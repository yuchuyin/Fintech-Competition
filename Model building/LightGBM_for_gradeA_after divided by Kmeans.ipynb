 # A _LABEL=0

```python
y_trainA=pd.read_csv(path+'/grade/A/label/y_trainA_L1.csv',index_col=0)
y_testA=pd.read_csv(path+'/grade/A/label/y_testA_L1.csv',index_col=0)

X_testA=pd.read_csv(path+'/grade/A/label/X_testA_L1.csv',index_col=0)
X_trainA=pd.read_csv(path+'/grade/A/label/X_trainA_L1.csv',index_col=0)
# 處理 y
le = LabelEncoder()
le.fit(y_trainA['loan_status'].unique())  
y_trainA['ans']=le.transform(y_trainA['loan_status'])
```

```python
X_testA.drop(columns=['Unnamed: 0.1'])
X_trainA.drop(columns=['Unnamed: 0.1'])
```

```python
X_trainA = X_trainA.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '',x))
X_testA = X_testA.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '',x))
```

# 1

```python
parameters = {
 'num_leaves':list(range(100,500,10)),
 'device': ['gpu']
}

gbm = lgb.LGBMClassifier(objective = 'binary',
                         is_unbalance = True,
                         metric = 'binary_logloss,auc',
                         max_depth = -1,
                         learning_rate = 0.1,
                         feature_fraction = 0.7,
                         min_child_samples=20,
                         min_child_weight=0.001,
                         bagging_fraction = 1,
                         bagging_freq = 2,
                         reg_alpha = 0.001,
                         reg_lambda = 8,
                         cat_smooth = 0,
                           
                        )

t0 = time.time()
gsearch = GridSearchCV(gbm, param_grid=parameters, scoring='f1', cv=10)
scores = cross_val_score(gbm, X_trainA,y_trainA.ans, cv=10, scoring='f1')
lgb_A=gsearch.fit(X_trainA, y_trainA.ans)


print(scores)
t1 = time.time()
print('gpu version elapse time: {}'.format(t1-t0))
```

```python
gsearch.best_params_
```

# KMEANS 

```python
y_trainA=pd.read_csv(path+'/grade/A/y_trainA.csv')
y_testA=pd.read_csv(path+'/grade/A/y_testA.csv')
X_testA=pd.read_csv(path+'/grade/A/X_testA.csv',index_col=0)
X_trainA=pd.read_csv(path+'/grade/A/X_trainA.csv',index_col=0)
```

```python
X_allA=pd.concat([X_trainA,X_testA],axis=0,ignore_index=True,verify_integrity=False)
```

```python
y_allA=pd.concat([y_trainA,y_testA],axis=0,ignore_index=True,verify_integrity=False)
```

```python
X_allA_ID=X_allA['Unnamed: 0.1']
```

```python
X_allA.drop(columns=['Unnamed: 0.1'])
```

```python
from sklearn.cluster import KMeans
from sklearn import cluster
from sklearn import cluster, datasets


# KMeans 演算法
kmeans_fit = KMeans(n_clusters=2,init='random', n_init=500, max_iter=1000, tol=0.0001, precompute_distances='auto', 
verbose=0, random_state=None, copy_x=True, n_jobs=1, algorithm='auto').fit(X_allA)

# 印出分群結果
cluster_labels = kmeans_fit.labels_
print("分群結果：")
print(cluster_labels)
print("---")


```

```python
# 加 回 'Unnamed: 0.1' 、 CUSTER LABEL 、 Y
```

```python
X_allA['Unnamed: 0.1']=X_allA_ID
X_allA['label']=cluster_labels
allA=pd.merge(X_allA,y_allA)
```

```python
# 按照 custer label 分成2個 dataframe
```

```python
grade_list=allA.label.unique().tolist()
group=allA.groupby('label')
print(group.size())

allA_L1=group.get_group(1)
allA_L0=group.get_group(0)
```

```python
# 分成 train test
```

```python
n_test_L1=allA_L1.loan_status.isnull().sum(axis=0)
print(n_test_L1)
n_test_L0=allA_L0.loan_status.isnull().sum(axis=0)
print(n_test_L0)


trainA_L1=allA_L1[:allA_L1.shape[0]-n_test_L1]
trainA_L0=allA_L0[:allA_L0.shape[0]-n_test_L0]

testA_L1=allA_L1[allA_L1.shape[0]-n_test_L1:allA_L1.shape[0]]
testA_L0=allA_L0[allA_L0.shape[0]-n_test_L0:allA_L0.shape[0]]
```

```python
# 分 X_train y_train  X-test  y_test
```

```python
X_trainA_L1=trainA_L1.drop(columns=['Unnamed: 0', 'loan_status','label'])
y_trainA_L1=trainA_L1[['Unnamed: 0.1','Unnamed: 0', 'loan_status']]

X_trainA_L0=trainA_L0.drop(columns=['Unnamed: 0', 'loan_status','label'])
y_trainA_L0=trainA_L0[['Unnamed: 0.1','Unnamed: 0', 'loan_status']]


X_testA_L1=testA_L1.drop(columns=['Unnamed: 0', 'loan_status','label'])
y_testA_L1=testA_L1[['Unnamed: 0.1','Unnamed: 0', 'loan_status']]

X_testA_L0=testA_L0.drop(columns=['Unnamed: 0', 'loan_status','label'])
y_testA_L0=testA_L0[['Unnamed: 0.1','Unnamed: 0', 'loan_status']]
```

```python
X_trainA_L1.to_csv(path+'/grade/A/label/X_trainA_L1.csv')
y_trainA_L1.to_csv(path+'/grade/A/label/y_trainA_L1.csv')

X_trainA_L0.to_csv(path+'/grade/A/label/X_trainA_L0.csv')
y_trainA_L0.to_csv(path+'/grade/A/label/y_trainA_L0.csv')


X_testA_L1.to_csv(path+'/grade/A/label/X_testA_L1.csv')
y_testA_L1.to_csv(path+'/grade/A/label/y_testA_L1.csv')

X_testA_L0.to_csv(path+'/grade/A/label/X_testA_L0.csv')
y_testA_L0.to_csv(path+'/grade/A/label/y_testA_L0.csv')
```
